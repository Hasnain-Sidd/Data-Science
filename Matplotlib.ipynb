```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
# line Plot 
# ms means marker size, mec marker edge color, mfc marker face color
n=256
x=np.linspace(-np.pi,np.pi,n,endpoint=True)
y=np.sin(2*x)
plt.plot(x,y,color='blue',alpha=1.00)
plt.fill_between(x,y,0,color='blue',alpha=.2)
plt.show()

```

```python
# Score of Students 
y=np.array([6,6,5,4,6,4,5,3,6,0,1,6,3,3,6,5])
x=np.array(['x1','x2','x3','x4','x5','x6','x7','x8','x9','x10','x11','x12','x13','x14','x15','x16'])
plt.plot(x,y,marker='o',color='blue',mfc='red',mec='blue',ms=7,linestyle='dashed')
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student's Scores")
plt.grid(color='black',linestyle='dashed')
plt.show()
```

```python
# Profit of Two companies Year Wise 
x=np.array([2001,2002,2003,2004,2005,2006,2007,2008])
puma=np.array([10,15,17,22,27,34,37,43])
adidas=np.array([8.2,12,19,25,29,36,40,47])
under_armour=np.array([6.4,9.6,15,29.5,31.7,38.2,46.7,58.32])
plt.plot(x,puma,marker='o',color='blue',mfc='blue',mec='blue',ms=7,linestyle='dashed')
plt.plot(x,adidas,marker='o',color='brown',mfc='brown',mec='brown',ms=7,linestyle='dashed')
plt.plot(x,under_armour,marker='o',color='purple',mfc='purple',mec='purple',ms=7,linestyle='dashed')
plt.xlabel("Year")
plt.ylabel("Profit $M")
plt.title("Profit of two Companies Year Wise")
plt.legend(labels=('Puma','Adidas','Under Armour'),loc='best',shadow=True)
plt.grid(color='black',linestyle='dashed',linewidth=1)
plt.show()
```

```python
# Match  scorecard
overs=np.array([5,10,15,20,25,30,35,40,45,50])
Ind=np.array([30,65,93,130,165,190,215,264,278,314])
Pak=np.array([19,43,80,118,149,210,240,285,307,320])
plt.plot(overs,Ind,color='blue',linestyle='solid')
plt.plot(overs,Pak,color='green',linestyle='dashed')
plt.xlabel("Overs")
plt.ylabel("Runs")
plt.title("Pak vs Ind")
plt.legend(labels=("India","Pakistan"),loc='best',title='Score',shadow=True)
plt.grid(color='black',linestyle="solid",linewidth=1,axis='y')
plt.show()
```

```python
# Bar Graph
menmeans=[20,34,49,38,57]
womenMeans=[25,31,36,49,51]
gr=np.arange(5)
plt.bar(gr,womenMeans,color='blue',width=.30)
plt.bar(gr-.30,menmeans,color='red',width=.30)
plt.xlabel("Groups")
plt.ylabel("Scores")
plt.xticks(gr-.15,('G1','G2','G3','G4','G5'))
plt.yticks(np.arange(0,60,3))
plt.legend(labels=('Women','Men'),loc='best',shadow=True)

plt.show()
```

```python
# Way OF Convice Used by Student to Raeech the School
x=['Walking','Cycling','Car','Bus','Train']
y=[29,15,35,18,3]
plt.bar(x,y,width=.65,color='green')
plt.grid(color='black',linestyle='solid',linewidth=.2,axis='y')
plt.xlabel("Way of Travel")
plt.ylabel("Student Reached School")
plt.title("Type of Transport Used By Student to Reach the School")
plt.show()
```

```python
# Number Of Apple And Banana Eat by Different Age Group People
app=[1,7,12,11,19]
ban=[2,8,14,20,34]
group=np.arange(5)
plt.bar(group,app,color='gray',width=.35)
plt.bar(group-.35,ban,color='black',width=0.35)
plt.xticks(group-0.2,('0-5','5-10','10-15','15-20','20+'))
plt.xlabel("Age")
plt.ylabel("Apples/Banana Eat")
plt.yticks(np.arange(0,40,3))
plt.legend(labels=('Apple','Banana'),loc='best',shadow=True)
plt.grid(color='black',linestyle='solid',linewidth=0.5,axis='y')
plt.title("Fruits Eat By Different Age Group")
plt.show()
```

```python
Pak=[54,51,64,62,88]
Eng=[75,63,42,50,83]
group=np.arange(5)
plt.bar(group,Pak,color='green',width=.40)
plt.bar(group-.40,Eng,color='blue',width=0.40)
plt.xticks(group-0.2,('1-10','11-20','21-30','31-40','41-50'))
plt.xlabel("Overs")
plt.ylabel("Runs")
plt.yticks(np.arange(0,100,10))
plt.legend(labels=('Pakistan','England'),loc='best',shadow=True,title='Scorecard')
plt.grid(color='black',linestyle='solid',linewidth=0.3,axis='y')
plt.title("Pakistan vs England Odi Scorecard")
plt.show()
```

```python
Sport=['Cricket','Football','Hockey','BasketBall','Pro Wrestling']
Watch_sport=np.array([73,85,46,29,19])
plt.pie(Watch_sport,autopct="%1.2f%%",shadow=True)
plt.title("Percentage Of People Watching Different Sports Across the World")
plt.legend(labels=Sport,shadow=True,loc='best',prop={'size':7})
plt.show()
```

```python
data=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
lang=['Java','Python','PHP','Javascript','C#',"C++"]
exp=[0.2,0,0,0,0,0.1]
plt.pie(data,labels=lang,autopct="%1.1f%%",explode=exp,shadow=True,startangle=120)
plt.title("Popularity of Programming Langauges\n WorldWide, Oct 2017 compared to year ago",color='Maroon')
plt.show()
```

```python
country=['United States','Great Britain','China','Russia','Germany']
gold_medals=[46,27,26,19,17]
exp=[0.2,0,0,0,0]
color=['blue','red','Maroon','orange','gray']
plt.pie(gold_medals,labels=country,autopct="%1.2f%%",explode=exp,shadow=True,startangle=120,colors=color)
plt.title('Gold Medal Achievements of Five Different Countries in 2016 Summer Olympics',color='Maroon')
plt.show()
```

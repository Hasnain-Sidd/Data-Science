```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
# line Plot 
# ms means marker size, mec marker edge color, mfc marker face color
n=256
x=np.linspace(-np.pi,np.pi,n,endpoint=True)
y=np.sin(2*x)
plt.plot(x,y,color='blue',alpha=1.00)
plt.fill_between(x,y,0,color='blue',alpha=.2)
plt.show()

```

```python
# Score of Students 
y=np.array([6,6,5,4,6,4,5,3,6,0,1,6,3,3,6,5])
x=np.array(['x1','x2','x3','x4','x5','x6','x7','x8','x9','x10','x11','x12','x13','x14','x15','x16'])
plt.plot(x,y,marker='o',color='blue',mfc='red',mec='blue',ms=7,linestyle='dashed')
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student's Scores")
plt.grid(color='black',linestyle='dashed')
plt.show()
```

```python
# Profit of Two companies Year Wise 
x=np.array([2001,2002,2003,2004,2005,2006,2007,2008])
puma=np.array([10,15,17,22,27,34,37,43])
adidas=np.array([8.2,12,19,25,29,36,40,47])
under_armour=np.array([6.4,9.6,15,29.5,31.7,38.2,46.7,58.32])
plt.plot(x,puma,marker='o',color='blue',mfc='blue',mec='blue',ms=7,linestyle='dashed')
plt.plot(x,adidas,marker='o',color='brown',mfc='brown',mec='brown',ms=7,linestyle='dashed')
plt.plot(x,under_armour,marker='o',color='purple',mfc='purple',mec='purple',ms=7,linestyle='dashed')
plt.xlabel("Year")
plt.ylabel("Profit $M")
plt.title("Profit of two Companies Year Wise")
plt.legend(labels=('Puma','Adidas','Under Armour'),loc='best',shadow=True)
plt.grid(color='black',linestyle='dashed',linewidth=1)
plt.show()
```

```python
overs=np.array([5,10,15,20,25,30,35,40,45,50])
Ind=np.array([30,65,93,130,165,190,215,264,278,314])
Pak=np.array([19,43,80,118,149,210,240,285,307,320])
plt.plot(overs,Ind,color='blue',linestyle='solid')
plt.plot(overs,Pak,color='green',linestyle='dashed')
plt.xlabel("Overs")
plt.ylabel("Runs")
plt.title("Pak vs Ind")
plt.legend(labels=("India","Pakistan"),loc='best',title='Score',shadow=True)
plt.grid(color='black',linestyle="solid",linewidth=1,axis='y')
plt.show()
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
def Conter_Plot():
     path="C:\\Users\\PMLS\\Desktop\\Data_Sets\\E-pharmacy- Addressing medication errors and preference over on-site pharmacy .csv"
     df=pd.read_csv(path)
     ax=sns.countplot(data=df,x=df['Have you ever purchased medicines online?'], hue=df['Have you ever purchased medicines online?'],palette=['red','green'],width=0.3)
     for bars in ax.containers:
      ax.bar_label(bars,color='black',fontsize=11)
     ax.legend(labels=['No','Yes'],title='Purchase Status',shadow=True,fontsize=8)
     ax.set_facecolor('lightblue')
     plt.grid(color='black',linestyle='solid',linewidth=0.6,axis='y')
     plt.xlabel("Have you ever purchased medicines online?",color='purple',fontsize=12)
     plt.ylabel('Count',color='purple',fontsize=12)
     plt.title("Survey Of People Buy Medicines Online",color='maroon',fontsize=15)
     plt.show()
Conter_Plot()
```

```python

path="C:\\Users\\PMLS\\Desktop\\Data_Sets\\E-pharmacy- Addressing medication errors and preference over on-site pharmacy .csv"
df=pd.read_csv(path)
fig=plt.figure(figsize=(15,5))
Labels=np.array(["Yes","No","Haven't purchased it online","Not take online medicines"])
counts = df["Have you ever experienced a medication error via online purchasing?"].value_counts()
exp=[0,0,0.1,0]
ordered_counts = [counts.get(label, 0) for label in Labels]
color={'red','blue','green','purple'}
plt.subplot(1,2,1)
plt.pie(ordered_counts,autopct="%0.2f%%",explode=exp,startangle=280,shadow=True,colors=color)
plt.title("People Buy Online Medicine",color='maroon',fontsize=15)
plt.legend(labels=Labels,loc='best',shadow=True,fontsize=7)
path="C:\\Users\\PMLS\\Desktop\\Data_Sets\\E-pharmacy- Addressing medication errors and preference over on-site pharmacy .csv"
df=pd.read_csv(path)
y=df["How satisfied are you with the online services provided by that pharmacy?"].value_counts()
plt.subplot(1,2,2)
plt.bar(y.index,y,label={'Satisfied','Not Satisfied','Very Satisfied'},color={'gray','black','green'},width=0.3)
plt.xlabel("Satisfaction Level",fontsize=15,color='maroon')
plt.ylabel('Count',fontsize=15,color='maroon')
plt.title("Satisfaction with Online Pharmacy",color='maroon',fontsize=15)
plt.legend(labels=['Satisfied','Not Satisfied','Very Satisfied'],loc='best',shadow=True)
plt.grid(color='black',linestyle='solid',linewidth=0.6,axis='y')
plt.show()

```

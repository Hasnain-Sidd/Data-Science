```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path="C:\\Users\\PMLS\\Downloads\\Customer Churn.csv"
df=pd.read_csv(path)
print(df.head())
```

```python
df.isnull().sum()
```

```python
# Question # 01
plt.bar(df['Age Group'],df['Frequency of SMS'],width=0.35)
plt.bar(df['Age Group']-0.35,df['Frequency of use'],width=0.35)
plt.xlabel("Age Group",fontsize=12,color='maroon')
plt.ylabel("Total Count",fontsize=12,color='maroon')
plt.title("Identify Group's Preferences Over SMS/Calls",fontsize=15,color='maroon')
plt.legend(labels=['SMS','Calls'],shadow=True,loc='best')
plt.show()
print("Age Group 2,3,& 4 send more SMS messages than Phone Calls")
```

```python
short_call_duration = 1000
medium_call_duration = 7000

def call_category(duration):
    if duration <= short_call_duration:
        return 'short'
    elif duration <= medium_call_duration:
        return 'medium'
    else:
        return 'long'

df['call_category'] = df['Seconds of Use'].apply(call_category)

# Group data by age and call category, count distinct calls
call_counts = df.groupby(['Age Group', 'call_category']).size().unstack(fill_value=0)

# Create a bar chart
call_counts.plot(kind='bar', stacked=True,color=['black','blue','red'])
plt.xlabel('Age Group',color='Maroon',fontsize=12)
plt.ylabel('Number of Calls',color="Maroon",fontsize=12)
plt.title('Number of Distinct Phone Calls by Age Group (Short, Medium, Long)',color='Maroon',fontsize=15)
plt.legend(title='Call Category',shadow=True)
plt.xticks(rotation=0)
plt.show()
```

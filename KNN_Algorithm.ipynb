```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report
import seaborn as sns
```

```python
iris=load_iris()
Data=iris.data
df=pd.DataFrame(Data)
Target=iris.target
df2=pd.DataFrame(Target)
Concate=pd.concat([df,df2],axis='columns')
Concate.columns=['Sepal Length(cm)','Sepal Width(cm)','Petal Length(cm)','Petal width(cm)','Target']
print(Concate.head())
```

```python
# Breaking the data frame into three data frame 
DF0=Concate[:50]
DF1=Concate[50:100]
DF2=Concate[100:]
```

```python
# Visualizing  Iris Data
plt.scatter(DF0['Sepal Width(cm)'],DF0['Sepal Length(cm)'],color='green',marker="o",label='Sentosa',s=60,alpha=0.7)
plt.scatter(DF1['Sepal Width(cm)'],DF1['Sepal Length(cm)'],color='red',marker="o",label='Versicolor',s=60,alpha=0.7)
plt.scatter(DF2['Sepal Width(cm)'],DF2['Sepal Length(cm)'],color='orange',marker="o",label='Verginica',s=60,alpha=0.7)
plt.xlabel("Sepal Width in (cm)",color='purple',fontsize=12)
plt.ylabel("Sepal Length in (cm)",color='purple',fontsize=12)
plt.legend(loc='best',shadow=True)
plt.grid(linestyle='solid',linewidth=0.5,color='black')
plt.title("Iris Data Visualization",color='purple',fontsize=15)
plt.show()
```

```python
# Spiliting the values for our machine learning model
X=Concate.drop('Target',axis='columns')
y=Concate.Target
X_train,X_test,Y_Train,Y_test=train_test_split(X,y,test_size=0.2,random_state=1)
print(X_train.size)
print(Y_Train.size)
```

```python
# Fitting the values in our model
knn=KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train,Y_Train)
```

```python
# Evaluating the score of our model
knn.score(X_test,Y_test)
```

```python
y_pred=knn.predict(X_test)
cm=confusion_matrix(Y_test,y_pred)
```

```python
# Plot a confusion Matrix diagram for evaluating the performance of our model 
plt.figure(figsize=(8,5))
sns.heatmap(cm,annot=True,cmap='viridis')
plt.xlabel("Predicted",color='maroon',fontsize=12)
plt.ylabel("Truth",color='maroon',fontsize=12)
plt.title("Confusion Matix",fontsize=15,color='maroon')
plt.show()
```

```python
# Making a classification report on our model
cr=classification_report(Y_test,y_pred)
print(cr)
```

```python
print(data_frame3.head())
```

```python
y1=data_frame3['Target']
X1=data_frame3.drop('Target',axis='columns')
X_train1,X_test1,Y_Train1,Y_test1=train_test_split(X1,y1,test_size=0.2,random_state=42)
neighbors=np.arange(1,10)
for neighbor in neighbors:
       knn1=KNeighborsClassifier(n_neighbors=neighbor)
       knn1.fit(X_train1,Y_Train1)
      
       
y_pred1=knn1.predict(X_test1)
print(y_pred1)
```

```python
knn1.score(X_test1,Y_test1)
```

```python
cm1=confusion_matrix(Y_test1,y_pred1)
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(cm1,annot=True,cmap='magma')
plt.xlabel("Predicted",fontsize=12,color='purple')
plt.ylabel("Truth",fontsize=12,color='purple')
plt.title("Confusion Matrix",fontsize=15,color='purple')
plt.show()
```

```python
cs1=classification_report(Y_test1,y_pred1)
print(cs1)
```
